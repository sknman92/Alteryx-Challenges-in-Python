```python
# importing pandas and regex

import pandas as pd

import re
```

```python
# importing data

df = pd.read_excel('.\data.xlsx')
```

```python
df.iloc[15:30]
```

```python
# tokenizing by genres

df['Genres'] = df['Genres'].apply(lambda x: re.split(r',\s*', x))
```

```python
# exploding df on genres

df = df.explode('Genres').reset_index(drop = True)

df
```

```python
# filtering to fiction and nonfiction genres

df_filtered = df.loc[(df['Genres'].str.contains(r'^fiction$', case = False)) |(df['Genres'].str.contains(r'^nonfiction$', case = False))]

df_filtered
```

```python
# uppercasing genre

df_filtered['Genres'] = df_filtered['Genres'].str.title()

df_filtered
```

```python
# creating function to select a random book by max page number and genre


genre = 'Fiction'
max_page = 400

def select_book(x):
    book_genre = x.loc[(x['Genres'] == genre)]
    book_page = book_genre.loc[df['Number of pages'] < max_page]
    book_random = book_page.sample(n=1)
    
    return book_random
    

# applying function
select_book(df_filtered)
```
