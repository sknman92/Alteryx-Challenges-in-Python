```python
# importing pandas and numpy

import pandas as pd
import numpy as np
```

```python
# importing data

df = pd.read_excel('./data.xlsx')
```

```python
df.head()
```

```python
# assign points to gold

conditions = [
                (df['Medal'] == 'Gold'),
                (df['Medal'] == 'Silver'),
                (df['Medal'] == 'Bronze')
]

choices = [3, 2, 1]

df['point'] = np.select(conditions, choices, default = 'Unknown')

df['point'] = pd.to_numeric(df['point'])

df.head()
```

```python
# filtering data to summer

df_filtered = df[df['Season'] == 'Summer']
```

```python
# finding the most successful sport per team

df_filtered.groupby(['Team', 'Sport'])['point'].sum().groupby(level = 0, group_keys = False).nlargest(1)
```

```python
# finding the most successful person per team

df_filtered.groupby(['Team', 'Name']).agg(point_sum=('point', 'sum'),
                                         event=('Event', ', '.join))\
.reset_index()\
.sort_values(by = 'point_sum', ascending = False)\
.groupby('Team').head(1)
```
