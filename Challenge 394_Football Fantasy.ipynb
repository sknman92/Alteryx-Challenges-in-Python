```python
# setting scaffold end

def round_end (x):
        
    if x== 6:
        return 1
    else:
        return 6

df_round['end'] = df_round['start'].apply(round_end)

df_round.head()
```

```python
# scaffolding based on start and end for each row

scaffold = []

for index, row in df_round.iterrows():
    
    draft_round = row['Draft Round']
    init = row['start']
    end = row['end']
    round_val = init
    
    if init == 6:
    
        while round_val >= end:
            
            row['round'] = round
            scaffold.append([init, end, round_val, draft_round])
            round_val -= 1
            
    else:
        
        while round_val <= end:
            
            row['round'] = round
            scaffold.append([init, end, round_val, draft_round])
            round_val += 1
        
```

```python
scaffold
```

```python
# converting scaffold in dataframe

df_scaffold = pd.DataFrame(scaffold, columns = ['start', 'end', 'team', 'draft_round'])

df_scaffold.head(10)
```

```python
# adding sequence to df scaffold

df_scaffold['sequence'] = range(1, len(df_scaffold) + 1)

df_scaffold.head(10)
```

```python
# filtering out players with DST role

players_filtered = df_players[~df_players['POS'].str.contains('DST')]

players_filtered.head()
```

```python
# parsing position without number

players_filtered['position'] = players_filtered['POS'].apply(lambda x: pd.Series(re.split(r'\d+', x)))[[0]]

players_filtered.head()
```

```python
# sort by position desc and RK asc

players_sorted = players_filtered.sort_values(by = ['position', 'RK'], ascending = [False, True])

players_sorted.head()
```

```python
# adding record id

players_sorted['id'] = players_sorted.groupby('position').cumcount() + 1

players_sorted.head()
```

```python
# filtering to id <= 6

players_id = players_sorted[players_sorted['id'] <=6]

players_id.head(10)
```

```python
# adding sequence

players_id['sequence'] = range(1, len(players_id) + 1)

players_id.head()
```

```python
# joining df scaffold with players_id

df_merge = pd.merge(players_id, df_scaffold, on = 'sequence')

df_merge.head()
```

```python
# parsing player name

df_merge['name'] = df_merge['PLAYER NAME'].apply(lambda x: pd.Series(re.findall(r'(.+?)\(', x)))

df_merge.head(10)
```
