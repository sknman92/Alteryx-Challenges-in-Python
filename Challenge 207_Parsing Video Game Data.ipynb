```python
# importing lib

import pandas as pd

import re

import os
```

```python
# setting display

pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000)  # Expand display width
pd.set_option('display.max_colwidth', None)  # Show full content in a cell
```

```python
# grabbing cwd

cwd = os.getcwd()

cwd
```

```python
# joining file name to path

file_name = 'data.xlsx'

file_path = os.path.join(cwd, file_name)

file_path
```

```python
# importing data

df = pd.read_excel(file_path)

df.head()
```

```python
# filtering out first row

header = df.iloc[[0]]

header
```

```python
# filtering out the rest of the rows

content = df.iloc[1:,:]

content.head(5)
```

```python
# splitting header on pipe

header_split = header['Video Game Data'].str.split('|', expand = True)

header_split
```

```python
content
```

```python
# parsing content

content['parsed'] = content['Video Game Data'].apply(lambda x: re.findall(r'^(\d+)' # game id
                                                      r'([^\(]+)' # game title
                                                      r'\((\d+)\) ' # game year
                                                      r'CAT:(.+?) ' # game category
                                                      r'PUB:(.+?) ' # game publisher
                                                      r'\$(.+?)million, ' # NA sales
                                                      r'\$(.+?)million, ' # EU sales
                                                      r'\$(.+?)million, ' # JP sales
                                                      r'\$(.+?)million, ' # other sales
                                                      r'\$(.+?)million' # global sales
                                                      , x))


content.head()
```

```python
# checking data type of column

content['parsed'].apply(type)
```

```python
# checking data type of element within list

content['parsed'].apply(lambda x: [type(i) for i in x])
```

```python
# converting tuple to list

content['parsed_list'] = content['parsed'].apply(lambda x: list(x[0]))        

content.head()
```

```python
content['parsed_list'].apply(type)
```

```python
# testing function to return all elements in parse_list as one list
   
parse_list = []

def list_parser(row):
    
    length = len(row)
    
    for i in range(length):
        
        parse_list.append(row[i])
        
# needs to be applied to each row via a loop

for index, row in content.iterrows():
    
    list_parser(row['parsed_list'])

list_length = (len(parse_list))

print(f"length: {list_length}")
```

```python
# splitting parsed list into columns

content_normalized = content['parsed_list'].apply(lambda x: pd.Series(x))

content_normalized.head()
```

```python
# extracting new header as dict

new_columns = header_split.iloc[0,0:]

col_dict = dict(new_columns)

col_dict
```

```python
# renaming columns

content_normalized.rename(columns = col_dict)
```
