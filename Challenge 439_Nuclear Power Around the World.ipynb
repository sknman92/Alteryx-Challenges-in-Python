# 1. Which country experienced the most significant decrease in nuclear capacity from one noted year to the next adjacent noted year?

```python
# extracting column names from first row

header = op_power.iloc[0, 1:].tolist()

header
```

```python
[op_power.columns[0]]
```

```python
# adding country column header to the rest of the columns

op_power.columns = [op_power.columns[0]] + header
```

```python
op_power.head()
```

```python
# want to adjust the column names so that NaN year is filled with previous value

header_list = op_power.columns[1:].tolist()

header_list
```

```python
new_header = []

previous_header = pd.NA

for index, header in enumerate(header_list):
    
    if pd.isna(header):
        new_header.append(previous_header)
    else:
        new_header.append(header)
        previous_header = header
```

```python
new_header
```

```python
# adding new_header to column

op_power.columns = [op_power.columns[0]] + new_header

op_power.head()
```

```python
# concating second row header to column header

# extracting column header

column_header = op_power.columns[1:].tolist()

# extracting values from second row

second_header = op_power.iloc[1, 1:]

# concating both

new_header = [f"{header} {value}" for header, value in zip(column_header, second_header)]

new_header
```

```python
# imputing new_header to column

op_power.columns = [op_power.columns[0]] + new_header

# truncating data to relevant rows and columns

op_power = op_power.iloc[2:36, :17].reset_index(drop = 'true')

op_power
```

```python
# transposing data

op_power_transposed = pd.melt(op_power, 
       id_vars = ['Country'],
       var_name = 'Name',
       value_name = 'Value')\
.sort_values(by = ['Country', 'Name'])

op_power_transposed
```

```python
# finding answer

# filtering data to just MW(e)

op_power_capacity = op_power_transposed[op_power_transposed['Name'].str.contains('MW\(e\)', case = False)]

op_power_capacity
```

```python
# changing value type to numeric and sorting

op_power_capacity['Value'] = op_power_capacity['Value'].astype('float')

op_power_capacity = op_power_capacity.sort_values(by=['Country', 'Name'], ascending = True)

op_power_capacity
```

```python
# creating function to find YoY value

def YoY_Change(country):
    
    results = []
    previous_value = pd.NA
    
    for index, row in country.iterrows():
        if pd.isna(previous_value):
            results.append(pd.NA)
        else: results.append(row['Value'] - previous_value)
        
        previous_value = row['Value']
        
    return pd.Series(results, index = country.index)

# applying function

YoY_Change = op_power_capacity.groupby('Country').apply(YoY_Change).reset_index(level = 0, drop = True) # don't quite understand 
# why I have to set level to 0

op_power_capacity['YoY_Change'] = YoY_Change
```

```python
# finding largest difference in YoY_Change

op_power_capacity.sort_values(by = ['Country', 'YoY_Change'], ascending = False).groupby('Country').head(1).sort_values(by = 'YoY_Change', ascending = False)
```

# 2. Which country saw the most significant drop in nuclear units from 1995 to 2022?

```python
# filtering data to No. of units

op_power_units = op_power_transposed[op_power_transposed['Name'].str.contains('No\.')]

op_power_units.head()
```

```python
# filtering data to 1995 and 2015

op_power_units_filtered = op_power_units[op_power_units['Name'].str.contains('1995') | op_power_units['Name'].str.contains('2022')]

op_power_units_filtered.head()

# changing data type of value to float

op_power_units_filtered['Value'] = op_power_units_filtered['Value'].astype('float')

op_power_units_filtered.head()
```

```python
# finding difference between 2022 and 1995

op_power_units_filtered['YoY_Change'] = op_power_units_filtered.groupby('Country')['Value'].diff()

# finding max YoY_Change to rid of nulls

op_power_units_filtered.groupby('Country')['YoY_Change'].max().reset_index().sort_values(by=['YoY_Change'])
```

# 3. Which country has the highest average percentage use of nuclear power?

```python
# slicing data to rid of total and misc rows

nuclear_prod = nuclear_prod.iloc[:36]

# getting rid of null columns

nuclear_prod = nuclear_prod.dropna(axis = 1, how = 'all')

nuclear_prod
```

```python
# extracting first row as column headers

column_headers = nuclear_prod.iloc[0, 1:].tolist()

column_headers
```

```python
# filling in nan with year

previous_row = pd.NA
result = []

for index, value in enumerate(column_headers):
    if pd.isna(value):
        result.append(previous_row)
    else: result.append(value)
        
    previous_row = value
        
result
```

```python
# extracting second row as column sub headers

column_sub_headers = nuclear_prod.iloc[1, 1:].tolist()

column_sub_headers
```

```python
# concating column_headers and column_sub_headers

column_headers = [f"{result} {column_sub_headers}" for result, column_sub_headers in zip(result, column_sub_headers)]

column_headers
```

```python
# imputing new columns to df

nuclear_prod.columns = [nuclear_prod.columns[0]] + column_headers

# shaping to revelant rows

nuclear_prod = nuclear_prod.iloc[2:]

nuclear_prod.head()
```

```python
# tranposing data

nuclear_prod_transposed = pd.melt(nuclear_prod,
       id_vars = 'Country',
       var_name = 'Name', 
       value_name = 'Value')

nuclear_prod_transposed
```

```python
# filtering data to % of total

nuclear_prod_transposed = nuclear_prod_transposed[nuclear_prod_transposed['Name'].str.contains('%')]

nuclear_prod_transposed.head()
```

```python
# changing percentage to 0 if null

nuclear_prod_transposed['Value'] = nuclear_prod_transposed['Value'].apply(lambda x: '0' if pd.isna(x) else x)

nuclear_prod_transposed.head()

# changing data type

nuclear_prod_transposed['Value'] = nuclear_prod_transposed['Value'].astype('float')

nuclear_prod_transposed.head()
```

```python
# finding average % of nuclear usage for countries

nuclear_prod_transposed.groupby('Country').agg(avg_usage = ('Value', 'mean'))\
.sort_values(by='avg_usage', ascending = False)
```

```python
# trying it a different way without filtering to % of nuclear usage by using a vectorized condition

nuclear_prod_transposed['Value'] = nuclear_prod_transposed['Value'].astype('float') # changing data type

nuclear_prod_transposed.groupby('Country').agg(avg_usage = ('Value', lambda x: x[nuclear_prod_transposed.loc[x.index, 'Name'].str.contains('%')].mean())).sort_values(by='avg_usage', ascending = False)
```
