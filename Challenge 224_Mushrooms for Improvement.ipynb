```python
# importing pandas and regex

import pandas as pd
import re
```

```python
# importing os and grabbing cwd

import os

cwd = os.getcwd()

cwd
```

```python
# joining path with file name

path = os.path.join(cwd, 'mushrooms.xlsx')

# grabbing mushroom data

df_mushrooms = pd.read_excel(path)

df_mushrooms.head()
```

```python
# grabbing attribute data

df_attribute = pd.read_excel(path, sheet_name = 'Sheet2')

df_attribute.head()
```

```python
# adding record id to mushrooms data

df_mushrooms['recordid'] = range(1, len(df_mushrooms) + 1)

df_mushrooms.head()
```

```python
# transposing

# selecting columns to pivot

pivot_cols = df_mushrooms.columns[0:23]

df_mushrooms_pivot = pd.melt(df_mushrooms, id_vars = ['recordid'], value_vars = pivot_cols)

df_mushrooms_pivot.head()
```

```python
# concating name and value

df_mushrooms_pivot['category-value'] = df_mushrooms_pivot['variable']+' '+(df_mushrooms_pivot['value'])

df_mushrooms_pivot.head()
```

```python
# trimming attribute Field1

df_attribute['Field1'] = df_attribute['Field1'].str.strip()

df_attribute.head()
```

```python
# replacing attribute information in first row 


df_attribute['Field1'] = df_attribute['Field1'].str.replace('Attribute Information: ', '')

df_attribute.head()
```

```python
# replacing '(' and ')'

df_attribute['Field1'] = df_attribute['Field1'].apply(lambda x: re.sub(r'[()]', '', x))

df_attribute.head()
```

```python
# splitting field to separate columns

# using just str split func with expand

# df_attribute['Field1'].str.split(r'[:()]+', expand=True)

# using lambda with re.split and pd.Series

df_attribute[['category', 'value']]= df_attribute['Field1'].apply(lambda x: pd.Series(re.split(r'[:]', x)))
df_attribute.head()
```

```python
# tokenizing value into rows

df_attribute['value'] = df_attribute['value'].str.split(',')
```

```python
df_attribute.head()
```

```python
# exploding into rows

df_attribute = df_attribute.explode('value')

df_attribute.head()
```

```python
# splitting value into two columns

# method 1

df_attribute[['value_header', 'value_code']] = df_attribute['value'].str.split('=', expand = True)

df_attribute.head()
```

```python
# method 2

df_attribute['value'].apply(lambda x: pd.Series(re.split('=', x)))
```

```python
# adjusting 'classes' category to class

df_attribute['category'] = df_attribute['category'].apply(lambda x: x if x != 'classes' else 'class')

df_attribute.head()
```

```python
# concating category and value

df_attribute['category-value'] = df_attribute['category'] + ' ' + df_attribute['value_code']

df_attribute.head()
```

```python
# mimicing find and replace on category-value from both tables to replace with value_header 

# first, creating dictionary of category-value and value in df_attribute

cat_value_dict = dict(zip(df_attribute['category-value'], df_attribute['value_header']))

cat_value_dict
```

```python
# creating function to replace category-value in df_mushrooms_pivot with value_header

df_mushrooms_pivot = df_mushrooms_pivot.replace(cat_value_dict)

df_mushrooms_pivot.head()
```

```python
# crosstabing df_mushrooms

df_final = pd.crosstab(df_mushrooms_pivot['recordid'], df_mushrooms_pivot['variable'], values = df_mushrooms_pivot['category-value'], aggfunc='min')
```
