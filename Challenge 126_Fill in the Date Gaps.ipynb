```python
# importing pandas and numpy

import pandas as pd

import numpy as np
```

```python
# importing data

df = pd.read_csv(r'C:\Users\CharlesYi\Jupyter Notebook\Alteryx Challenges\Challenge 126_Fill in the Date Gaps\data.csv')

df.head()
```

```python
# returning unique countries

country_array = df['country_name'].unique()

country_array
```

```python
# converting array into df

df_country = pd.DataFrame(country_array, columns = ['country'])

df_country
```

```python
# scaffolding according to start date of 2017-10-01 and end date of 2018-07-23


date_range = pd.date_range(start = '2017-10-01', end = '2018-07-23', freq = 'D')

country_dates = []

for country in df_country['country']:
    
    for x in date_range:
        
        line = {'country' : country, 'date' : x}
        
        country_dates.append(line)
```

```python
country_dates[:20]
```

```python
# turning list into df

df_country_dates = pd.DataFrame(country_dates)

df_country_dates.head()
```

```python
# need to convert dtype of date in original df to datetime to match dtype in df_country_dates

df['datetime_finished'] = pd.to_datetime(df['datetime_finished'])
```

```python
# joining country_dates to original df

df_merged = pd.merge(df, df_country_dates, left_on = ['country_name', 'datetime_finished'], right_on = ['country', 'date'], how = 'right')

df_merged.head()
```

```python
# cleaning up columns

df_merged = df_merged[['date', 'country', 'passed_exams']]

df_merged
```

```python
# checking to make sure passed exams came through

df_merged[df_merged['passed_exams'] == 1].head()
```

```python
# filling na in passed exams with 0

df_merged['passed_exams'] = df_merged['passed_exams'].fillna(0)
```

```python
# calculating running total of passed exams


df_merged['runningtot'] = df_merged.groupby(['country'])['passed_exams'].transform('cumsum')

df_merged.tail(100)
```
