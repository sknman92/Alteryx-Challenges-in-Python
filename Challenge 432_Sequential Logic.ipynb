```python
# importing pandas

import pandas as pd
```

```python
# creating df

dict = {"A" : [1, 1, 2],
        "B" : [0, 1, 1],
        "C" : [0, 0, 0]}

df = pd.DataFrame(dict)
```

```python
# Adding additional rows
for i in range(22):
       df.loc[len(df)] = [
                           df.iloc[-1].sum(), 
                           df.iloc[[-1], 0:3].apply(lambda x: x['A'] - x['B'], axis = 1).iloc[0], 
                           df.iloc[[-1], 0:3].apply(lambda x: x['A'] - (x['B'] + x['C']), axis = 1).iloc[0]
                        ]

df
```
