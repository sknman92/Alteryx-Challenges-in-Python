```python
final_combined_table = pd.DataFrame()

for x in manager_list:
    
    df_branch = df_merged[df_merged['Assistant_To_Regional_Manager'] == x]
    
    # summ'ing first and last date for each branch

    date_summ = df_branch.groupby('Bank Branch ID').agg(first_date = ('Transaction DateTime', 'min'), last_date = ('Transaction DateTime', 'max'))

    date_summ.reset_index(inplace = True)
    
    # returning first and last of month

    date_summ['first_date'] = date_summ['first_date'].dt.to_period('M').dt.start_time.dt.strftime('%Y-%m-%d')
    date_summ['last_date'] = date_summ['last_date'].dt.to_period('M').dt.end_time.dt.strftime('%Y-%m-%d')
    
    # converting dates to date type

    date_summ['first_date'] = pd.to_datetime(date_summ['first_date'])

    date_summ['last_date'] = pd.to_datetime(date_summ['last_date'])
    
    # iterating rows to scaffold

    dict_list = []

    for index, row in date_summ.iterrows():
    
        date = row['first_date']
        last_date = row['last_date']
        bank_branch_id = row['Bank Branch ID']
    
        while date <= last_date:
        
            data_dict = {'Bank Branch ID' : bank_branch_id,
                         'date_scaffold' : date}
        
            dict_list.append(data_dict)
            
            date += pd.DateOffset(days = 1)
        
    date_scaffold = pd.DataFrame(dict_list)
    
    date_scaffold['exchanged_transaction_amount'] = 0
    
    # unioning transaction amount to scaffold

    df_sliced = df_branch[['Bank Branch ID', 'Exchanged Transaction Amount', 'Transaction DateTime']] # grabbing relevant columns

    df_sliced.columns = ['Bank Branch ID', 'exchanged_transaction_amount', 'date_scaffold'] # renaming columns

    df_unioned = pd.concat([date_scaffold, df_sliced]) # unioning
    
    # summarizing transaction amount by branch id and date

    df_trans_summ = df_unioned.groupby(['Bank Branch ID', 'date_scaffold']).agg(x_trans_amount = ('exchanged_transaction_amount', 'sum')).reset_index()
    
    # including total sum amout for each bank branch

    df_opening_balance = df_branch.groupby(['Bank Branch ID', 'Assistant_To_Regional_Manager']).agg(opening_balance = ('Exchanged Opening Balance' , 'sum')).reset_index()
    
    # joining trans sum and opening balance on bank branch id

    trans_balance_merged = df_trans_summ.merge(df_opening_balance, on = 'Bank Branch ID')
    
    # sorting by bank branch ID and date

    trans_balance_merged.sort_values(by = ['Bank Branch ID', 'date_scaffold'], inplace = True)
    
    trans_balance_merged.reset_index(drop = True, inplace = True)
    
    # creating function to find daily balance

    def multi_row(group):
    
        daily_balance_list = []

        previous_balance = None
    
        for _, row in group.iterrows():
    
            opening_balance = row['opening_balance']
    
            if pd.isna(previous_balance):
                daily_balance = row['opening_balance']
                daily_balance_list.append(daily_balance)
        
            else:
                daily_balance = row['x_trans_amount'] + previous_balance
                daily_balance_list.append(daily_balance)
        
            previous_balance = daily_balance
            
        return pd.Series(daily_balance_list, index=group.index)
    
    # applying function
    
    daily_balance_series = trans_balance_merged.groupby('Bank Branch ID').apply(multi_row).reset_index().iloc[:, 1]
    
    trans_balance_merged['daily_balance'] = daily_balance_series
    
    # creating debit and credit columns

    trans_balance_merged['debit'] = trans_balance_merged['x_trans_amount'].apply(lambda x: x if x > 0 else 0)

    trans_balance_merged['credit'] = trans_balance_merged['x_trans_amount'].apply(lambda x: x if x < 0 else 0)
    
    # aggregating debit, credit, and avg daily balance

    agg_branch = trans_balance_merged.groupby(['Bank Branch ID', 'Assistant_To_Regional_Manager']).agg(debit = ('debit', 'sum'), credit = ('credit', 'sum'),
                                                                                     avg_daily_balance = ('daily_balance', 'mean'),
                                                                                     opening_balance = ('opening_balance', 'max'))

    agg_branch.reset_index(inplace=True)
    
    # aggregating total debit, credit, and avg daily balance for regional manager

    agg_manager = agg_branch.groupby('Assistant_To_Regional_Manager').agg(opening_balance = ('opening_balance', 'sum'),
                                                        debit = ('debit', 'sum'),
                                                        credit = ('credit', 'sum'),
                                                        avg_daily_balance = ('avg_daily_balance', 'mean')
                                                        )


    agg_manager.reset_index(inplace=True)

    # unioning agg at branch and manager levels

    final_table = pd.concat([agg_branch, agg_manager])
    
    # calculating ending balance

    final_table['ending_balance'] = final_table['opening_balance'] + (final_table['debit'] + final_table['credit'])
    
    # calculating % change

    final_table['%_change'] = ((final_table['ending_balance'] - final_table['opening_balance']) / final_table['opening_balance'])*100
    
    # selecting relevant columns

    final_table = final_table[['Bank Branch ID', 'Assistant_To_Regional_Manager', 'opening_balance', 'ending_balance', 'debit', 'credit', 'avg_daily_balance', '%_change']]

    # concat'ing final table
    
    final_combined_table = pd.concat([final_combined_table, final_table])
    
```
