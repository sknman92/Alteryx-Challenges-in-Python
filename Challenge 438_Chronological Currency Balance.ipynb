```python
# importing data

import pandas as pd

df = pd.read_excel('./data.xlsx')
```

```python
df
```

```python
# transposing
 
df_melted = pd.melt(df,
        id_vars = ['Subgroup Name', 'Order', 'Currency'],
       var_name = 'Date',
       value_name = 'Balance')
```

```python
# converting to datetime

df_melted['Date'] = pd.to_datetime(df_melted['Date'])
```

```python
df_melted_sorted = df_melted.sort_values(by = ['Order', 'Date'], ascending = True)

df_melted_sorted
```

```python
# pivoting back to original structure

pd.pivot(df_melted_sorted,
        index = ['Subgroup Name', 'Order', 'Currency'],
        columns = 'Date',
        values = ['Balance'])
```
