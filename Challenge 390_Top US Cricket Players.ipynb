Challenge Objective: Assemble Your Dream US Cricket Team 

Your mission—should you choose to accept it—is to identify the crème de la crème of US run scorers in cricket. You will be handed a treasure trove of international cricket match data, and your goal is to sift through the numbers to pinpoint the Top 5 US run scorers who would be the perfect fit for your dream team. 

```python
# importing lib

import pandas as pd

import os

import re
```

```python
# importing data

cwd = os.getcwd()

file_name = 'data.csv'

file_path = os.path.join(cwd, file_name)

df = pd.read_csv(file_path)

df.head()
```

```python
df.dtypes
```

```python
# examining whether runs col has characters other than digits

df[df['Runs'].str.match(r'[^\d]+')]
```

```python
# since runs col has strings in it, replace strings with 0

df['Runs'] = [x if re.match(r'[\d]+', x) else '0' for x in df['Runs']]
```

```python
df.head()
```

```python
# examining whether runs col has characters other than digits

df[df['Runs'].str.contains(r'\D', regex=True)]'

# several rows have * appended to number of runs
```

```python
# creating another list comprehension to eliminate *

df['Runs'] = [x if re.match(r'^\d+$', x) else x.replace('*', '') for x in df['Runs']]
```

```python
df.head()
```

```python
# converting runs to numerica

df['Runs'] = df['Runs'].astype('int16')
```

```python
df.head()
```

```python
# filtering team to USA

df_filtered = df[df['Team'] == 'U.S.A.']

df_filtered.head()
```

```python
# aggregating number of runs by player

df_agg = df_filtered.groupby('Player_Name')['Runs'].sum().reset_index()

df_agg.head()
```

```python
# selecting players with the 5 highest runs

df_agg.nlargest(5, 'Runs')
```
