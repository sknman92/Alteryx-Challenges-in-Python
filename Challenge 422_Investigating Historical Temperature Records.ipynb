```python
# importing pandas

import pandas as pd
```

```python
# importing data

df = pd.read_csv('./data.csv')
```

```python
# converting dt field to date time

df['dt'] = pd.to_datetime(df['dt'])

df.dtypes
```

```python
# filtering data to 1950 onwards

df_filtered = df.loc[df['dt'] >= '1950-01-01']

df_filtered.shape
```

1. Determine which cities have average temperatures greater than or equal to 25 degrees.

```python
df_avgtemp = df_filtered.groupby(['Country', 'City'])['AverageTemperature'].mean().reset_index(name = 'AvgTemp').query('AvgTemp >= 25').sort_values(by = 'AvgTemp')

df_avgtemp
```

2. Among the cities identified in the previous task, identify the country with the highest number of such cities.

```python
df_avgtemp.groupby('Country').size().reset_index(name = 'number_cities').nlargest(1, 'number_cities')
```

3. Examining all countries within the dataset, pinpoint the year with the highest average temperature and the year with the lowest average temperature across the globe.

```python
# creating year column

df['year'] = df['dt'].dt.year

df
```

```python
avgtemp_year = df.groupby('year')['AverageTemperature'].mean().reset_index(name = 'avgtempyear').round({'avgtempyear': 2})

avgtemp_year
```

```python
# filtering to 1950 onwards

avgtemp_year = avgtemp_year[avgtemp_year['year'] >= 1950]

avgtemp_year
```

```python
def highlow(df, year, avgtempyear):
    
    min_temp = round(df['avgtempyear'].min(), 2)
    max_temp = round(df['avgtempyear'].max(), 2)
    
    min_year = df.loc[df['avgtempyear'] == min_temp].reset_index(drop = True).rename(columns = {'year' : 'min_year', 'avgtempyear' : 'min_temp'})
    max_year = df.loc[df['avgtempyear'] == max_temp].reset_index(drop = True).rename(columns = {'year' : 'max_year', 'avgtempyear' : 'max_temp'})
    
    frames = [min_year, max_year]
    
    return pd.concat(frames, axis = 1)
        
```

```python
highlow(avgtemp_year, avgtemp_year['year'], avgtemp_year['avgtempyear'])
```

```python
# could have used idx function instead

def highlow(df, year, avgtempyear):
    
    min_year = df.loc[df['avgtempyear'].idxmin()].to_frame().T.reset_index(drop = True).rename(columns = {'year' : 'min_year', 'avgtempyear' : 'min_temp'})
    max_year = df.loc[df['avgtempyear'].idxmax()].to_frame().T.reset_index(drop = True).rename(columns = {'year' : 'max_year', 'avgtempyear' : 'max_temp'})
    
    frames = [min_year, max_year]
    
    return pd.concat(frames, axis = 1)
```

```python
highlow(avgtemp_year, avgtemp_year['year'], avgtemp_year['avgtempyear'])
```
