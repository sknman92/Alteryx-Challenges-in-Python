```python
# importing regex and numpy

import numpy as np
import re

# converting columns to str

df_test = df.iloc[:, 1:].astype('str')
```

```python
df_trans = pd.concat([df.iloc[:, 0], df_test], axis = 1)
```

```python
# cleaning up df to convert to integer

pattern = r'\d+'

def regex(x):
    if re.match(pattern, x):
        return x
    else:
        return np.nan
    
# df_trans.iloc[:, 1:].applymap(lambda x: np.where(re.match(pattern, x)), 0, np.nan) -- dont understand still why this doesnt work

df_trans_2 = df_trans.map(regex)

df_trans_2
```

```python
# converting strings back to integer

df_test = df_trans_2.iloc[:, 1:].astype('float64')

df_trans_2 = pd.concat([df_trans_2.iloc[:, 0], df_test], axis = 1)
```

```python
# summing rows to find total

df_trans_2['total'] = df_trans_2.apply(lambda x: x[1:].sum(), axis = 1)
```

```python
# finding percentage

def perc_calc(row):
    data_filtered = row.iloc[1:48]
    total = row['total']
    perc = data_filtered/total
    header = (row.iloc[[0, 50]])
    return pd.concat([header, perc])

df_trans_2.apply(perc_calc, axis = 1)
```
