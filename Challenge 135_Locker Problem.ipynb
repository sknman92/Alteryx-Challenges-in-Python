```python
# importing lib

import pandas as pd
```

```python
# creating data

student_list = [i for i in range(1, 1001)]
```

```python
# converting to df

df = pd.DataFrame(student_list, columns = ['student_id'])

df.head()
```

```python
df.shape
```

```python
# creating function to assign locker number

def locker_number(df):
    
    lock_num_list = []
    
    for index, row in df.iterrows():
        
        iteration_num = 1
        
        while iteration_num <= 1000:
            
            new_row = row.copy()
        
            lock_num = new_row['student_id'] * iteration_num
            
            new_row['lock_num'] = lock_num

            lock_num_list.append(new_row)

            iteration_num += 1
        
    return lock_num_list
        
# this version was really slow...
```

```python
# faster version..

def locker_number(df):
    
    lock_num_list = []
    
    for index, row in df.iterrows():
        
        for i in range(1, len(df) + 1):
            
            lock_num = row['student_id'] * i
            
            new_row = {'student_id': row['student_id'], 'lock_num' : lock_num}
            
            lock_num_list.append(new_row)
        
    return lock_num_list
```

```python
locker_iterated = locker_number(df)
```

```python
df_iterated = pd.DataFrame(locker_iterated)

df_iterated.head()
```

```python
# filtering out where locker # greater than 1000

df_filtered = df_iterated[df_iterated['lock_num'] <= 1000]

df_filtered.head()
```

```python
# counting locker occurence by locker num

df_count = df_filtered.groupby('lock_num')['lock_num'].count().reset_index(name = 'count')

df_count.head()
```

```python
# adding status to each locker (open or closed)

df_count['status'] = df_count['count'].apply(lambda x: 'closed' if x % 2 == 0 else 'open')
```

```python
df_count
```

```python
# filtering to lockers that are open

df_final = df_count[df_count['status'] == 'open']

df_final.head()
```
