```python
# importing pd

import pandas as pd
```

```python
# creating df

df = pd.DataFrame({
                    "Amount" : [720, 680, 510, 550, 1000]
                })

df
```

```python
# iterating to scaffold to 4 weeks per amount

Weeks = []

for index, row in df.iterrows():
    Week = 1
    Amount = row['Amount']
    while Week <= 4:
        Weeks.append({"Amount" : Amount, "Week" : Week})
        Week += 1

Weeks
```

```python
# converting into df

df = pd.DataFrame(Weeks)

df
```

```python
# creating deficit amount calculating function

def deficit_calc(group):
    results = []
    previous_row = None
    
    for index, row in group.iterrows():
        if row['Week'] == 1:
            results.append(row['Amount'])
        elif row['Week'] != 1:
            deficit = results[-1] - 150
            results.append(deficit)
        
    group['Deficit'] = results
    
    return group
```

```python
# applying function

df = df.groupby('Amount').apply(deficit_calc).reset_index(drop = True)

df
```

```python
# importing numpy

import numpy as np
```

```python
df['Deficit'] = df.apply(lambda x: np.where(x['Deficit'] >= 150, 150, x['Deficit']), axis = 1)

df
```

```python
# crosstabbing

df_crosstab = pd.crosstab(df['Amount'], df['Week'], values = df['Deficit'], aggfunc = 'sum')

df_crosstab
```

```python
# finding carryover

def carryover_calc(row):
    total_available = row.name
    total_spent = row.sum()
    carryover = total_available - total_spent
    return carryover

    
df_crosstab['carryover'] = df_crosstab.apply(carryover_calc, axis = 1)

df_crosstab
     
```
