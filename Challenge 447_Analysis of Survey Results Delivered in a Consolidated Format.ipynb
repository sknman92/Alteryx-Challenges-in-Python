```python
# import pandas

import pandas as pd
```

```python
# importing questionnaire and questions list

df_q = pd.read_excel('./questionnaire.xlsx')

df_q_list = pd.read_excel('./Question_List.xlsx')
```

```python
# filtering df_q to completed surveys

df_q = df_q.loc[df_q['Status'] == 'Complete']
```

```python
# converting Time started and Date submitted into dt

df_q['Time Started'] = pd.to_datetime(df_q['Time Started'])

df_q['Date Submitted'] = pd.to_datetime(df_q['Date Submitted'])

df_q.head()
```

```python
# finding difference between time started and date submitted

from datetime import datetime

df_q['diff_minutes'] = (df_q['Date Submitted'] - df_q['Time Started']).dt.total_seconds() / 60

df_q
```

```python
# finding min, max, and median minutes

df_q.agg(min_minutes = ('diff_minutes', 'min'), max_minutes = ('diff_minutes', 'max'), median_mintues = ('diff_minutes', 'median'))
```

```python
# tokenizing question and answer on new line

df_q['Question_and_Answer'] = df_q['Question_and_Answer'].str.split(r'\n')

# exploding question and answer

df_q = df_q.explode('Question_and_Answer')
```

```python
# assigning 1 to question on jupyter notebook frequency and answer

def record(x):

    record_id = []
    q = None
    
    for index, row in x.iterrows():
        if q == '- How often do you use Jupyter Notebook?' and '?' not in row['Question_and_Answer']:
            record_id.append(1)
        elif row['Question_and_Answer'] == '- How often do you use Jupyter Notebook?':
            record_id.append(1)
        else: record_id.append(0)
        
        q = row['Question_and_Answer']
    
    return record_id

df_q['record'] = record(df_q)
            
    
```

```python
df_q.head()
```

```python
# filtering to record 1 and answer

df_q_filtered = df_q.loc[(df_q['record'] == 1) &  (~df_q['Question_and_Answer'].str.contains(r'\?'))]
```

```python
# creating function to sum total answers then find percentage of each answer to total

def answer_breakdown(df):
    
    total_answers = df.shape[0]
    
    # find number of each answers
    
    answer_count = df.groupby('Question_and_Answer').agg({'Question_and_Answer' : 'count'})
    
    # appending total_answers to answer_count
    
    answer_count['percentage'] = (answer_count['Question_and_Answer'] / total_answers) * 100
    
    return answer_count
    
answer_breakdown(df_q_filtered)
    
```
